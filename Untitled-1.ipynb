```python
import random
```

```python
wuxia_characters = ["Lin Lanfei",
"Yi Aoxin",
"Wu Zhijian",
"Lan Bingren",
"Ming Yuzhai",
"Da Deming",
"Angjian",
"Jie Hedian",
"Jie Aibing",
"Wu Xiongxin",
"Lan Anying",
"Qing Shi",
"Yong Jie",
"Baiye Yuyan",
"Zhao Renshu",
"Yin Ying"
]

keys_characters = [
    "Neris Dionte",
    "Beiei",
    "Edelweiss Van Aalsburg",
    "Kali Eir",
    "Leila Maurer",
    "Edwin Rhys",
    "Hari Layette",
    "Lerofey Attin",
    "Linyi",
    "Cian",
    "Headmistress",
    "Alice Van Aalsburg",
    "Alexi Van Aalsburg",
    "General Eleos",
    "Mika Dionte",
    "Thalos Korus",
    "the sunik",
    "Nestor Draco",
    "Ash Belrose",
    "Mio Robin",
    "Clio Amarette",
    "the fairy",
    "Dolly",
    "Elchik on Tov"
]
```

```python
choice = input("Choose a character list - Wuxia (1),  Keys (2), Both (3) : ")

if choice == "1":
    pool = wuxia_characters
elif choice == "2":    
    pool = keys_characters
elif choice == "3":
    pool = wuxia_characters + keys_characters
else:
    print("Invalid choice. Defaulting to Both.")
```

```python
gold_star_input = input("Include gold star percentage? (y/n): ")

if gold_star_input.lower() == 'y':
    gold_star_character = None
    
    # Give user 3 attempts
    for attempt in range(3):
        gold_star_character = input("Enter gold star character name: ")
        
        if gold_star_character in pool:
            # Valid character, get percentage
            try:
                gold_star_percentage = float(input("Enter gold star percentage (e.g., 5 for +5%): "))
                gold_star_percentage = gold_star_percentage / 100  # Convert to decimal
            except ValueError:
                print("Invalid percentage. Defaulting to 0.")
                gold_star_percentage = 0
            break
        else:
            print(f"'{gold_star_character}' not found in pool.")
            if attempt < 2:  # Not the last attempt
                print(f"Attempts remaining: {2 - attempt}")
            else:
                print("Max attempts reached. Continuing without gold star boost.")
                gold_star_character = None
                gold_star_percentage = 0
else:
    gold_star_character = None
    gold_star_percentage = 0

print(gold_star_character, gold_star_percentage)
```

```python
base_weight = 1
weights = [base_weight] * len(pool)

if gold_star_character and gold_star_character in pool:
    gold_star_index = pool.index(gold_star_character)
    
    # Add the percentage as extra weight
    # Base probability for each character is 1/pool_size
    # Adding gold_star_percentage means adding that % to the base probability
    base_probability = 1 / len(pool)
    boosted_probability = base_probability + gold_star_percentage
    
    # Convert back to weight relative to others so that I can use random.choices
    weights[gold_star_index] = boosted_probability / base_probability

result = random.choices(pool, weights=weights, k=1)[0]
print(f"\nYou got: {result}")
```
